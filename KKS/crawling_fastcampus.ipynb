```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# webdriver 연결
browser = webdriver.Chrome()
```

```python
# 네이버블로그 패스트캠퍼스 후기 검색 페이지 연결
url = 'https://section.blog.naver.com/Search/Post.naver?pageNo=1&rangeType=PERIOD&orderBy=sim&startDate=2022-01-01&endDate=2022-12-31&keyword=%ED%8C%A8%EC%8A%A4%ED%8A%B8%EC%BA%A0%ED%8D%BC%EC%8A%A4%20%ED%9B%84%EA%B8%B0'
browser.get(url)
```

```python
# 페이지 연결 element 가져오기
pages_elements = browser.find_element(By.CLASS_NAME, 'pagination').find_elements(By.CLASS_NAME, 'item')

len(pages_elements)
```

```python
# 네이버 블로그에서 pagenum의 페이지에 해당하는 리뷰 크롤링
pagenum = 500
contents_list = []
for i in range(pagenum):
    # if idx > 0:
    #     break

    # page.click()
    browser.get(f'https://section.blog.naver.com/Search/Post.naver?pageNo={i+1}&rangeType=PERIOD&orderBy=sim&startDate=2022-01-01&endDate=2022-12-31&keyword=%ED%8C%A8%EC%8A%A4%ED%8A%B8%EC%BA%A0%ED%8D%BC%EC%8A%A4%20%ED%9B%84%EA%B8%B0')
    time.sleep(1)

    links_element = browser.find_elements(By.CLASS_NAME, 'desc_inner')
    links_review = []
    for link in links_element:
        links_review.append(link.get_attribute('href'))
        print(link.get_attribute('href'))

    contents = ""
    for idx, review in enumerate(links_review):
        # if idx > 1:
        #     break
        response = requests.get(review)
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        # print(soup)
        res_src = requests.get('https://blog.naver.com' + soup.find('iframe')['src'])
        print('https://blog.naver.com' + soup.find('iframe')['src'])
        html_src = res_src.text
        soup_src = BeautifulSoup(html_src, 'html.parser')
        # print(soup_src)

        span_elements = soup_src.find_all('span')
        for span in span_elements:
            contents += " " + span.get_text().replace('\n', '').replace('\t', '')
    
        contents_list.append(contents)
        contents = ""

# print(contents_list)
df = pd.DataFrame(contents_list)
df
```
