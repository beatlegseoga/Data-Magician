```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# 파일로드
df = pd.read_csv("eda-proj-fc-purchase.csv")
df
```

```python
# 테이블 정보
df.info()
```

```python
# 수치데이터 통계량
df.describe()
```

```python
# 강의별 매출액 top 20
df_top_sales = df.groupby('코스(상품) 이름')['실거래금액'].sum().sort_values(ascending=False).head(20)
df_top_sales.plot.bar()
```

```python
# x축, y축 변환 위한 데이터프레임 생성
df_top_sales = pd.DataFrame(df_top_sales)
df_top_sales
```

```python
# column 지정을 위한 인덱스 리셋
df_top_sales = df_top_sales.reset_index()
df_top_sales
```

```python
# seaborn 이용한 매출액 top 20 목록 시각화
plt.figure(figsize=(8,6))
plt.xticks(np.arange(0, 400000001, 100000000), labels=['0','1억','2억','3억','4억'])
sns.barplot(data=df_top_sales, x='실거래금액', y='코스(상품) 이름')
plt.show()
```
